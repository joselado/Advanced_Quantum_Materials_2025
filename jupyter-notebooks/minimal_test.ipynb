## One dimensional chain
First, we will consider a one dimensional chain

```python
from pyqula import geometry
g = geometry.chain() # generate a honeycomb lattice
gs = g.get_supercell(12) # make a supercell and plot the structure
plt.scatter(gs.r[:,0],gs.r[:,1],c="black",s=400) ; plt.axis("equal") ; plt.axis("off") 
```

```python
h = g.get_hamiltonian(tij=[1.0])
(k,e) = h.get_bands() # compute band structure
plt.scatter(k,e,c="black") ; plt.xlabel("Momentum") ; plt.ylabel("Energy") 
plt.xticks([]) ; plt.xlim([min(k),max(k)])


plt.tight_layout()
```
